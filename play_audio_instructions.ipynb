```python
# import libraries
import numpy as np
import pyaudio

# creat a test signal
Fs = 44100; dur = 3; freq = 500
sine = np.sin(2*np.pi*freq*np.linspace(0,dur,Fs*dur))
# multiply by 32000 to make it nearly fullscale (range of integer)
sine32 = sine * 32000

# make an instance of a player
p = pyaudio.PyAudio()

# prepare a stream with given format, channels and fs
stream = p.open(format=pyaudio.paInt16, channels=1, rate=Fs, output=True)

# start to write bytes to stream 
stream.write(sine32.astype('int16').tobytes())
```

```python
# show that integer make it fullscale 
# (or try and listen to the very quiet unscaled sine array)

import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(18,3))
plt.subplot(121); plt.plot(sine[0:1000]);
plt.subplot(122); plt.plot(sine32[0:1000]);
```
